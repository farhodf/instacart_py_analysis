## Step 6

```python
# Grouping customers and checking how much on average they spent per item
ords_prods_merge1['spend_average'] = ords_prods_merge1.groupby(["user_id"])['prices'].transform(np.mean)
```

```python
#creating the labesls
ords_prods_merge1.loc[ords_prods_merge1['spend_average'] >= 10 , 'spender_flag'] = "High Spender"
```

```python
ords_prods_merge1.loc[ords_prods_merge1['spend_average'] < 10 , 'spender_flag'] = "Low Spender"
```

```python
ords_prods_merge1['spender_flag'].value_counts(dropna = False)
```

## Step 7

```python
# Grouping customers and chcking how often they order 
ords_prods_merge1['order_frequency'] = ords_prods_merge1.groupby(["user_id"])['days_since_prior_order'].transform(np.median)
```

```python
#creating the labels
ords_prods_merge1.loc[ords_prods_merge1['order_frequency'] > 20 , 'frequency_flag'] = "Non-frequent customer"
```

```python
ords_prods_merge1.loc[(ords_prods_merge1['order_frequency'] <= 20) & (ords_prods_merge1['order_frequency'] > 10), "frequency_flag"] = 'Regular customer'
```

```python
ords_prods_merge1.loc[ords_prods_merge1['order_frequency'] <= 10 , 'frequency_flag'] = "Frequent customer"
```

```python
ords_prods_merge1['frequency_flag'].value_counts(dropna = False)
```
